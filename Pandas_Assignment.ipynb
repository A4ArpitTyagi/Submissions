```python
import pandas as pd
import numpy as np
```

```python
url = r'E:\AI Lead Tech\Pandas\flights.csv'
# load the data
flight = pd.read_csv(url)
flight.head()
```

```python
## Finding the number of canceled flights for every airline per weekday

# Group by needs to be used on airline and weekday and total count of cancelled
# flights need to be calculated after that.

################ SOLUTION 1 ##############################
# Step 1 : Filter on CANCELLED column for the value 1 as we need to know the count
# of cancelled flights

flight1=flight[flight['CANCELLED']==1]

# Step 2: Group by on above dataframe on AIRLINE and WEEKDAY as asked in the quesiton.
# and count on CANCELLED column.

flight1.groupby(['AIRLINE','WEEKDAY'])["CANCELLED"].count()
```

```python
## Question 2: Finding the number and percentage of cancelled and diverted flights
## for every airline per weekday

## Answer  - Part 1- Diverted Flights number and % numbers for the same.

flight.head()

# flight_delayed : Only those Rows where flights got diverted i.e. DIVERTED > 0
flight_delayed = flight[flight['DIVERTED']>0]


# Diverted SUM: using above DataFrame to know the count.
count_diverted = flight_delayed.groupby(['AIRLINE','WEEKDAY']).agg ({'DIVERTED':'count'})
#print(count_diverted)

#Filling 0 in place of NaN vaues
count_diverted['DIVERTED']=count_diverted['DIVERTED'].fillna(0)

# Total count to know the % Diverted flights
count_total = flight.groupby(['AIRLINE','WEEKDAY']).agg({'DIVERTED':'count'})
#print(count_total.head())

# Dividing two DataFrames to know the %.
percent_division = count_diverted.div(count_total)*100

# Filling 0 for all NaN values in the Diverted column
percent_division['DIVERTED'] = percent_division['DIVERTED'].fillna(0)

# Showing % sign in the 5 values and 2 decimal points.
pd.set_option('display.float_format','{:.2f}%'.format)

# Renaming the column name to show the % Diverted
percent_division.rename(columns={'DIVERTED': '% DIVERTED'},inplace=True)
#print(percent_division)

# Adding one more column to show the numerical value of the diverted flights too.
percent_division['DIVERTED'] =  count_diverted['DIVERTED']
# Filling 0 in place of NaN.
percent_division['DIVERTED'] = percent_division['DIVERTED'].fillna(0)

# Converting float to int by .apply(int) function.
percent_division['DIVERTED'] = percent_division['DIVERTED'].apply(int)

# Reordering of both the columns - First number then %.
percent_division = percent_division[['DIVERTED','% DIVERTED']]

print(percent_division)

```

```python
## Answer 2 - Part 2- Cancelled Flights number and % numbers for the same.

flight.head()

# flight_can : Only those Rows where flights got diverted i.e. DIVERTED > 0
flight_can = flight[flight['CANCELLED']>0]
# Cancelled SUM: using above DataFrame to know the count.
count_can = flight_can.groupby(['AIRLINE','WEEKDAY']).agg ({'CANCELLED':'count'})
#print(count_can)
#Filling 0 in place of NaN vaues
count_can['CANCELLED']=count_can['CANCELLED'].fillna(0)

# Total count to know the % Diverted flights
count_total = flight.groupby(['AIRLINE','WEEKDAY']).agg({'CANCELLED':'count'})
#print(count_total)
# Dividing two DataFrames to know the %.
percent_division = count_can.div(count_total)*100
# Filling 0 for all NaN values in the Diverted column
percent_division['CANCELLED'] = percent_division['CANCELLED'].fillna(0)
# Showing % sign in the 5 values and 2 decimal points.
pd.set_option('display.float_format','{:.2f}%'.format)
# Renaming the column name to show the % Diverted
percent_division.rename(columns={'CANCELLED': '% CANCELLED'},inplace=True)
#print(percent_division)
# Adding one more column to show the numerical value of the diverted flights too.
percent_division['CANCELLED'] =  count_can['CANCELLED']
# Filling 0 in place of NaN.
percent_division['CANCELLED'] = percent_division['CANCELLED'].fillna(0)
# Converting float to int by .apply(int) function.
percent_division['CANCELLED'] = percent_division['CANCELLED'].apply(int)

# Reordering of both the columns - First number then %.
percent_division = percent_division[['CANCELLED','% CANCELLED']]

print(percent_division)

```
